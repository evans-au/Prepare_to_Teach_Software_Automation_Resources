# Matplotlib

Matplotlib is a comprehensive library for creating data visualisations in Python. In machine learning, it is essential for:
- **Exploratory Data Analysis (EDA)**: Understanding data distributions, identifying outliers, and spotting patterns before training models.
- **Model Evaluation**: Visualising loss curves, accuracy metrics, confusion matrices, and decision boundaries.
- **Feature Engineering**: Plotting feature correlations, histograms, and scatter plots to inform feature selection.
- **Communicating Results**: Creating publication-quality figures to present findings.

```python
# Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Print the different inbuilt plot styles
print(plt.style.available)

# Apply the `ggplot` style
plt.style.use('ggplot')
```

#### Basic Plot Types
This example demonstrates three common plot types on the same figure: scatter plots for individual data points, line plots for trends, and bar charts for categorical comparisons.

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

# Scatter the data points
plt.scatter(x, y, marker='o', s=75, c='r',label='Values', zorder=3)
# Line the data points
plt.plot(x, y, marker='x', c='b',label='Trend', zorder=2)
# Bar plot the data
plt.bar(x, y, width=1, color='green', edgecolor="white", zorder=0)
# Set the title
plt.title("A Random Visualisation")
# Set the y-axis label
plt.ylabel('Target')
# Set the x-axis label
plt.xlabel('Feature')
plt.legend()
plt.show()
```

#### Plots as Objects (Vertical Layout)
Using `plt.subplots()` returns figure and axes objects, giving you more control over your visualisation. This example creates a vertical stack of three different plot types: scatter, line, and bar charts.

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
col = [1,2,3,4,5]
col_hex = ["#000000", '#ff0000', "#00ff00", "#0000ff", "#ffffff"]

fig,ax = plt.subplots(3)
ax[0].scatter(x, y, s=40, c=col)
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

ax[1].plot(x, y, marker='x', label='Trend')
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")

ax[2].bar(x, y, width=1, edgecolor="white", color=col_hex)
ax[2].set_xlabel("x")
ax[2].set_ylabel("y")

fig.suptitle("2x2 Grid of ax objects", fontsize=16)
#plt.tight_layout(rect=[0, 0, 1, 0.95]) 
plt.show()
```

#### Plots as Objects (Grid Layout)
This example demonstrates a 2x2 grid of subplots, useful for comparing multiple visualisations side by side. Use `plt.tight_layout()` to prevent overlapping labels.

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
col = [1,2,3,4,5]

fig,ax = plt.subplots(2,2)
ax[0,0].scatter(x, y, s=40, c=col)
ax[0,0].set_xlabel("x")
ax[0,0].set_ylabel("y")

ax[1,0].scatter(x, y, s=40, c=col)
ax[1,0].set_xlabel("x")
ax[1,0].set_ylabel("y")

ax[0,1].scatter(x, y, s=40, c=col)
ax[0,1].set_xlabel("x")
ax[0,1].set_ylabel("y")

ax[1,1].scatter(x, y, s=40, c=col)
ax[1,1].set_xlabel("x")
ax[1,1].set_ylabel("y")
fig.suptitle("2x2 Grid of ax objects", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95]) 
plt.show()
```

#### Heatmap
A heatmap displays data as a colour-encoded matrix, useful for visualising correlations, distributions, or any 2D data. The `imshow()` function renders the grid, and a colorbar provides a legend for the values.

```python
# Create sample data for heatmap
data = np.random.rand(5, 5)

fig, ax = plt.subplots()
heatmap = ax.imshow(data, cmap='viridis')

# Add colorbar
plt.colorbar(heatmap)

# Add labels
ax.set_xticks(np.arange(5))
ax.set_yticks(np.arange(5))
ax.set_xticklabels(['A', 'B', 'C', 'D', 'E'])
ax.set_yticklabels(['1', '2', '3', '4', '5'])

# Add values to cells
for i in range(5):
    for j in range(5):
        ax.text(j, i, f'{data[i, j]:.2f}', ha='center', va='center', color='white')

ax.set_title("Heatmap Example")
plt.show()
```
